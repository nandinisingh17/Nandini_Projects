```python
# .shift(), we can use the shift function so that we can compare yesterday's revenue to today's revenue

coffee['yesterday_revenue'] = coffee['Revenue'].shift(2) # doing 2 since there are two types of coffee
```

```python
coffee
```

```python
# we can then calculate the percentage of revenue increase in revenue from yesterday to today

coffee['pct_change'] = coffee['Revenue']/ coffee['yesterday_revenue']
```

```python
coffee
```

```python
# using the rank function to get the highedt rank of the player's height
bios['height_rank'] = bios['height_cm'].rank(ascending = False)
bios.sort_values(['height_rank'], ascending = False)
```

```python
# using the rolling function we can get the revenue of the past three days

latte = coffee[coffee['Coffee Type'] == "Latte"].copy()
latte ['3day']=latte['Units Sold'].rolling(3).sum()
```
